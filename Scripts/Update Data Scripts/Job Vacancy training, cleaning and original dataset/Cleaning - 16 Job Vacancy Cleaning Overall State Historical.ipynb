## Reformat Final Dataframe

```python
state_all_hist = ACT_NT_QLD_SA_TAS_VIC_WA_NSW_AUST_joined.copy()
```

```python
state_all_hist
```

```python
state_all_hist = state_all_hist.rename(columns={'OCCUPATIONS': 'occname', 'Time' : 'time'})
```

```python
state_all_hist.head(5)
```

```python
reformat_state_time = pd.DataFrame(state_all_hist.time.str.split('-').tolist(),
                                   columns = ['day','month','year'])
```

```python
reformat_state_time.head(5)
```

```python
state_all_hist['year'] = reformat_state_time.year
state_all_hist['month'] = reformat_state_time.month
state_all_hist['day'] = reformat_state_time.day
```

```python
state_all_hist = state_all_hist.drop(['time'], axis=1)
```

```python
state_all_hist
```

```python
state_all_hist = state_all_hist[["occname","year","month","day", "vacancy ACT", "vacancy NT", "vacancy QLD", "vacancy SA",
                       "vacancy TAS","vacancy VIC","vacancy WA","vacancy NSW","vacancy AUST"]]
```

```python
state_all_hist.head(10)
```

```python
state_all_hist.tail(10)
```
